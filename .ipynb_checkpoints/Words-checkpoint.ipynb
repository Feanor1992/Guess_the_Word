Processing dataset with english words from Kaggle(https://www.kaggle.com/code/jacoporepossi/wordle-a-python-implementation/data) and russion words from dictionery

```python
import pandas as pd
```

```python
raw_english_words = pd.read_csv('raw_english_words.csv')
raw_english_words['word'] = raw_english_words['word'].astype('str')
raw_english_words.head(5)
```

```python
raw_russian_words = pd.read_csv('raw_russian_word.txt', encoding='utf-8')
raw_russian_words['word'] = raw_russian_words['word'].astype('str')
raw_russian_words.head(5)
```

Choosing the most common English words from dataset (occurs more then 1000000 counts)

```python
english_words = raw_english_words[(raw_english_words['count'] > 1000000)]
english_words
```

Choosing words with 5 letters from english words and russian words datafreims

```python
mask = english_words['word'].str.len() == 5
english_words_5_letters = english_words.loc[mask]
english_words_5_letters
```

```python
mask1 = raw_russian_words['word'].str.len() == 5
russian_words = raw_russian_words.loc[mask1]
russian_words
```

```python
del english_words_5_letters['count']
```

```python
english_words_5_letters.to_csv('english_words.csv')
russian_words.to_csv('russian_words.txt', encoding='utf-8')
```
